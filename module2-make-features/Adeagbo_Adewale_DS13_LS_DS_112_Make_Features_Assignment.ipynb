```python
df = pd.read_csv('LoanStats_2018Q4.csv', skiprows=1,skipfooter=2)
df.head()
```

```python
df = df.drop(['url','member_id','desc','id'], axis=1)
df.head()
```

```python
def int_rate_to_float(cell_content):
  return float(cell_content.strip().strip('%'))
```

```python
df['int_rate_float'] = df['int_rate'].apply(int_rate_to_float)
df.head()
```

```python
df['int_rate_float'].dtype
```

```python
# function to clean up emp_title
def clean_title(title):
  if isinstance(title, str):
    return title.strip().title()
  else:
    return "Unknown"
  
```

```python
df['emp_title'] = df['emp_title'].apply(clean_title)
df.head()
```

```python
df['issue_d'] = pd.to_datetime(df['issue_d'], infer_datetime_format=True)
df.dtypes[:15]
```

```python
df['issue_month'] = df['issue_d'].dt.month 
df['issue_year'] = df['issue_d'].dt.year
```

```python
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], infer_datetime_format=True)
```

```python
(df['issue_d'] - df['earliest_cr_line']).dt.days
```

```python
df['days_from_credit_to_issue'] = (df['issue_d'] - df['earliest_cr_line']).dt.days
```

```python
df['days_from_credit_to_issue'].describe()
```

```python
df['term'].value_counts(dropna=False)
```

```python
# converting term column from string to integer
def clean_term(term_cell):
  return int(term_cell.strip().strip('months'))
```

```python
df['term'] = df['term'].apply(clean_term)
df.head()
```

```python
# Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.
```

```python
df['loan_status'].value_counts(dropna=False)
```

```python
def loanStatus(cell):
  if cell == "Current" or cell == "Fully Paid":
    return 1
  else:
    return 0
```

```python
df['loan_status'] = df['loan_status'].apply(loanStatus)
```

```python
df['loan_status'].value_counts(dropna=False)
```

```python
# Make last_pymnt_d_month and last_pymnt_d_year columns.

df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)
```

```python
df['last_pymnt_d'].head(10)
```

```python
df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month 
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year 
df.head()
```
